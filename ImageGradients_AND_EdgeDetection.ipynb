```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

```

```python
img = cv2.imread("messi.jpeg", cv2.IMREAD_GRAYSCALE)

lap = cv2.Laplacian(img,cv2.CV_64F,ksize=3) #64 float is used to deal with negative numbers as we converted rgb to grayscale
lap = np.uint8(np.absolute(lap))   #converting into unsigned 8 bit integer

sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)  #1 denotes that we want to use sobelX method  (dX,dY) values are (1,0) for sobelX
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)  #(0,1) -- dX,dY
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv2.bitwise_or(sobelX,sobelY)



titles = ['original image', 'Laplacian', 'SobelX', 'SobelY', 'Sobel Combined']
images = [img, lap, sobelX, sobelY, sobelCombined]
plt.figure(figsize=(15,15))
for i in range(5):
    plt.subplot(3,2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
