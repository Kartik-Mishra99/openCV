```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("opencvlogo.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5),np.float32)/25  #dividing by width and height of kernel   (like CNN)
print("KERNEL :\n",kernel)
print("\n")
dst = cv2.filter2D(img,-1,kernel)  #applying homogenous filter --- destination image(dst)   #we are using filter2d to blur the image
blur = cv2.blur(img,(5,5))  #(5,5) is the kernel
gaussian_blur = cv2.GaussianBlur(img,(5,5),0)   # 0 is sigmaX value  --- read the docs.
median = cv2.medianBlur(img,3)   #kernel size must be odd --- used to work on salt and pepper noise
bilateral = cv2.bilateralFilter(img,9,75,75)  #preserves the edges very well (borders)

titles = ['original Image','2D Convolution','blur','Gaussian Blur','median','bilateral Filter']
images = [img,dst,blur,gaussian_blur,median,bilateral]

plt.figure(figsize=(15,15))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
